**Assume you have a dataframe df.Randomly select k centroids from df.**

```python
centroids=np.array(df.sample(n=3,random_state=42))
centroids
```

**Define a function with input as dataframe and centroids,and return the corresponding datapoint centroid labels.**

```python
def closest_centroid(df_array, centroids):
    return np.argmin(np.array([np.linalg.norm(df_array - centroid,axis=1) for centroid in centroids]).T,axis=1)
```

**Define function K_means_clustering which take dataframe,number of cluster,iteration as input and return final label and centroids.**

```python
def k_means_clustering(df, k=3, iterations=100):
  df_copy=df.copy()
  centroids=np.array(df_copy.sample(n=k,random_state=42))
  for _ in range(iterations):
    L=closest_centroid(df_copy,centroids=centroids)
    centroids=np.array([np.mean(df_copy[L==i],axis=0) for i in set(L)])
  return centroids,L

L,centroids=k_means_clustering(df,k=3,iterations=2)
print(L)
print(centroids)
```
